```python
# prompt: give me the cod to show the all unique types of the column suicide date. and show three example of each types. also show the count of each type write tis code as reuseabel function

import pandas as pd
def analyze_unique_types(df, column_name, num_examples=3):
    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in the DataFrame.")
        return

    unique_types = df[column_name].apply(type).unique()
    print(f"Unique data types in column '{column_name}':")
    for dtype in unique_types:
        print(f"- {dtype}")

    print(f"\nAnalysis of unique values and their types in column '{column_name}':")
    # Use .apply(type) to get the type of each entry, then value_counts()
    type_counts = df[column_name].apply(type).value_counts()

    for dtype, count in type_counts.items():
        print(f"\nData type: {dtype}")
        print(f"Count: {count}")
        print(f"Examples ({min(num_examples, count)}):")

        # Filter rows by the current data type and get examples
        examples = df[df[column_name].apply(type) == dtype][column_name].head(num_examples).tolist()
        for example in examples:
            print(f"  - {example}")
```

```python
import datetime

def convert_all_to_str(date):
    if isinstance(date, datetime.datetime):
        return date.strftime("%d/%m/%Y")
    elif isinstance(date, str):
        return date  # Already a string, leave it as is
    else:
        return None  # Covers NaN or any other types
```

```python
import re
import pandas as pd

def analyze_date_format(df, date_column, num_examples=3):
    if date_column not in df.columns:
        print(f"Error: Column '{date_column}' not found in the DataFrame.")
        return

    # Initialize counters and example collectors
    has_4digit_year = []
    no_4digit_year = []
    empty_or_null = []

    # Loop through the column with index
    for idx, val in df[date_column].items():
        if pd.isna(val) or str(val).strip() == "" or val is None:
            empty_or_null.append((idx, val))
        elif re.search(r'\b\d{4}\b', str(val)):
            has_4digit_year.append((idx, val))
        else:
            no_4digit_year.append((idx, val))

    # Print summary
    print(f"--- Analysis of column '{date_column}' ---")
    print(f"✅ Entries with a 4-digit year: {len(has_4digit_year)}")
    print(f"  Examples ({min(num_examples, len(has_4digit_year))}):")
    for idx, val in has_4digit_year[:num_examples]:
        print(f"    - Index {idx}: {val}")

    print(f"\n❌ Entries without a 4-digit year: {len(no_4digit_year)}")
    print(f"  Examples ({min(num_examples, len(no_4digit_year))}):")
    for idx, val in no_4digit_year[:num_examples]:
        print(f"    - Index {idx}: {val}")

    print(f"\n⚠️ Empty / Null / NaN entries: {len(empty_or_null)}")
    print(f"  Examples ({min(num_examples, len(empty_or_null))}):")
    for idx, val in empty_or_null[:num_examples]:
        print(f"    - Index {idx}: {val}")
    print("-" * (len(f"--- Analysis of column '{date_column}' ---")))
```

```python
analyze_date_format(df_clr, 'suicide_date')
```

```python
def fix_two_digit_year(date_str):
    if pd.isna(date_str) or not isinstance(date_str, str) or re.search(r'\b\d{4}\b', date_str):
        return date_str  # Already has a 4-digit year or is invalid/null

    parts = date_str.strip().split('/')
    if len(parts) == 3 and len(parts[2]) == 2:
        parts[2] = '20' + parts[2]
        return '/'.join(parts)
    return date_str  # Leave untouched if not in expected format
```

```python
# Apply to all rows
df_clr['suicide_date'] = df_clr['suicide_date'].apply(fix_two_digit_year)
```

```python
def print_unique_values(df, column_name):
  if column_name in df.columns:
    unique_values = df[column_name].unique()
    print(f"Unique values in column '{column_name}':")
    print(unique_values)
  else:
    print(f"Column '{column_name}' not found in the DataFrame.")
```

```python
def normalize_column_values(df, mapping_dict, column_name):
    return df[column_name].apply(
        lambda x: mapping_dict.get(str(x).strip().lower(), x) if pd.notna(x) else None
    )
```

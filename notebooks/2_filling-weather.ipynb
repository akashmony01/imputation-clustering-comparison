```python
def missing_data_report(df):
    # Calculate the percentage of missing data for each column
    missing_percentage = df.isna().mean() * 100

    # Create a DataFrame for the missing data summary
    missing_data_summary = pd.DataFrame({
        'Column': missing_percentage.index,
        'Missing Percentage': missing_percentage.values
    })

    # Define the four groups based on missing percentage
    group_1 = missing_data_summary[missing_data_summary['Missing Percentage'] > 70][['Column', 'Missing Percentage']]
    group_2 = missing_data_summary[(missing_data_summary['Missing Percentage'] > 40) & (missing_data_summary['Missing Percentage'] <= 70)][['Column', 'Missing Percentage']]
    group_3 = missing_data_summary[(missing_data_summary['Missing Percentage'] > 20) & (missing_data_summary['Missing Percentage'] <= 40)][['Column', 'Missing Percentage']]
    group_4 = missing_data_summary[missing_data_summary['Missing Percentage'] <= 20][['Column', 'Missing Percentage']]

    # Sort each group by missing percentage in descending order (most missing to least missing)
    group_1 = group_1.sort_values(by='Missing Percentage', ascending=False).reset_index(drop=True)
    group_2 = group_2.sort_values(by='Missing Percentage', ascending=False).reset_index(drop=True)
    group_3 = group_3.sort_values(by='Missing Percentage', ascending=False).reset_index(drop=True)
    group_4 = group_4.sort_values(by='Missing Percentage', ascending=False).reset_index(drop=True)


    # Rename columns for clarity
    group_1.columns = ['Column', 'Missing %']
    group_2.columns = ['Column', 'Missing %']
    group_3.columns = ['Column', 'Missing %']
    group_4.columns = ['Column', 'Missing %']

    # Display each group one by one, from most missing to least missing
    print("Group 1: Columns with >70% missing data")
    print(group_1.to_string(index=False))
    print("\nGroup 2: Columns with >40% to ≤70% missing data")
    print(group_2.to_string(index=False))
    print("\nGroup 3: Columns with >20% to ≤40% missing data")
    print(group_3.to_string(index=False))
    print("\nGroup 4: Columns with ≤20% missing data")
    print(group_4.to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from datetime import datetime

# Mapping of time of day to representative clock time
time_to_hour = {
    'morning': '08:00',
    'noon': '12:00',
    'afternoon': '15:00',
    'evening': '18:00',
    'night': '21:00',
    'day': '10:00'
}

# Step 1: Create a new column 'time_hour' from 'time' using the mapping
df_lat_lon['time_hour'] = df_lat_lon['time'].map(time_to_hour)

# Step 2: Define function to estimate unix_time from date and time_hour
def estimate_unix(row):
    date_str = row['suicide_date']
    time_str = row['time_hour']

    if pd.isna(date_str) or pd.isna(time_str):
        return row['unix_time']  # leave as-is

    try:
        dt = datetime.strptime(f"{date_str} {time_str}", "%d/%m/%Y %H:%M")
        return int(dt.timestamp())
    except Exception:
        return row['unix_time']  # leave as-is if parsing fails
```

```python
df_lat_lon['unix_time'] = df_lat_lon.apply(
    lambda row: estimate_unix(row) if pd.isnull(row['unix_time']) else row['unix_time'],
    axis=1
)
```

```python
import pandas as pd
from meteostat import Point, Hourly, Daily
from datetime import datetime, timedelta
from tqdm import tqdm
import pytz
```

```python
dhaka_tz = pytz.timezone("Asia/Dhaka")

def unix_to_dt(unix_time):
    try:
        utc_dt = datetime.utcfromtimestamp(unix_time)
        utc_dt = pytz.UTC.localize(utc_dt)
        dhaka_dt = utc_dt.astimezone(dhaka_tz)
        return dhaka_dt
    except Exception:
        return pd.NaT

# Weather columns to fill
weather_cols = ['temp_min', 'temp_max', 'air_pressure', 'air_humidity',
                'wind_speed', 'wind_deg', 'coco']
```

```python
for col in weather_cols:
    if col not in df_lat_lon.columns:
        df_lat_lon[col] = pd.NA
```

```python
def fill_weather(row):
    if all(pd.notna(row[col]) for col in ['temp_min', 'temp_max', 'air_pressure', 'air_humidity', 'wind_speed', 'wind_deg']):
        return row

    lat, lon, unix_time = row['latitude'], row['longitude'], row['unix_time']

    if pd.isna(lat) or pd.isna(lon) or pd.isna(unix_time):
        return row

    dt = unix_to_dt(unix_time)
    if pd.isna(dt):
        return row

    point = Point(lat, lon)

    # Try hourly data first
    try:
        from_dt = dt - timedelta(hours=1)
        to_dt = dt + timedelta(hours=1)
        hourly = Hourly(point, from_dt, to_dt)
        hourly_data = hourly.fetch()

        if not hourly_data.empty:
            if hourly_data.index.tz is None:
                hourly_data.index = hourly_data.index.tz_localize('UTC').tz_convert('Asia/Dhaka')
            else:
                hourly_data = hourly_data.tz_convert('Asia/Dhaka')

            time_diffs = hourly_data.index - dt
            closest_idx = time_diffs.abs().argmin()
            closest = hourly_data.iloc[closest_idx]

            fill_map = {
                'temp_min': 'temp',
                'temp_max': 'temp',
                'air_pressure': 'pres',
                'air_humidity': 'rhum',
                'wind_speed': 'wspd',
                'wind_deg': 'wdir',
                'coco': 'coco'
            }
            for col, key in fill_map.items():
                if pd.isna(row[col]) and key in closest:
                    val = closest.get(key)
                    if pd.notna(val):
                        row[col] = val
            return row
    except Exception:
        pass  # Silent, don't print per-row errors

    # Fallback to daily data
    try:
        date_only = pd.Timestamp(dt.date())
        daily = Daily(point, date_only, date_only)
        daily_data = daily.fetch()

        if not daily_data.empty:
            daily_row = daily_data.iloc[0]
            fill_map = {
                'temp_min': 'tmin',
                'temp_max': 'tmax',
                'air_pressure': 'pres',
                'air_humidity': 'rhum',
                'wind_speed': 'wspd',
                'wind_deg': 'wdir',
                'coco': 'coco'
            }
            for col, key in fill_map.items():
                if pd.isna(row[col]) and key in daily_row:
                    val = daily_row.get(key)
                    if pd.notna(val):
                        row[col] = val
    except Exception:
        pass  # Silent, don't print per-row errors

    return row
```

```python
tqdm.pandas(desc="Filling Weather Data")
df_lat_lon = df_lat_lon.progress_apply(fill_weather, axis=1)
```
